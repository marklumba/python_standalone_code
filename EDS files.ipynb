```python
# Import neccessary modules
import pandas as pd
import requests
import os
import xlwings as xw
```

```python
# URL's of the excel files in a dictionary
urls = {'Curt_EDS': 'Put your URL        'Aries_EDS':'Put your URL        'Luverne_EDS':'Put your URL        }
```

```python
# Iterate through urls and download the excel files
for file_name, url in urls.items():
    
    # Send a HTTP request to the URL
    response = requests.get(url)
    
    print(f"Downloading {url}")
    
    if response.status_code == 200:

      # Full path of the file
      path = os.path.join(os.path.expanduser("~"), "Desktop", f"{file_name}.xlsx")

      # Save the content of the response to a file on your desktop
      with open(path, 'wb') as f:
        f.write(response.content)
```

```python
# Full path of the file and naming convention
path = os.path.join(os.path.expanduser("~"), "Desktop", "Curt_EDS.xlsx")
path_2 = os.path.join(os.path.expanduser("~"), "Desktop", "Aries_EDS.xlsx")
path_3 = os.path.join(os.path.expanduser("~"), "Desktop", "Luverne_EDS.xlsx")

# Names of the sheets per excel file
sheet_names = ['Active Items', 'Digital Assets', 'Application Data']
sheet_names_1 = ['Active Item', 'Digital Assets', 'Application Data']
sheet_names_2 = ['Product Information', 'Digital Assets', 'Fitment']

# Read each sheet into a separate DataFrame df1, df2 and df3
df1 = pd.read_excel(path, sheet_name=sheet_names)
df2 = pd.read_excel(path_2, sheet_name=sheet_names_1)
df3 = pd.read_excel(path_3, sheet_name=sheet_names_2)
```

```python
# Sort the DataFrame df1['Application Data'] by 'Year' in descending order 
df1['Application Data'] = df1['Application Data'].sort_values(by=['Year', 'Part #'], ascending=[False, True])

# Drop duplicates and keeps the first occurence (which will be the one with the lastest year dye to the previous sorting)
df1['Application Data'] = df1['Application Data'].drop_duplicates(subset='Part #')

# Display the 5 rows
df1['Application Data'].head(5)
```

```python
# Sort the DataFrame df2['Application Data'] by 'Year' in descending order 
df2['Application Data'] = df2['Application Data'].sort_values(by=['Year', 'Part #'], ascending=[False, True])

# Drop duplicates and keeps the first occurence (which will be the one with the lastest year dye to the previous sorting)
df2['Application Data'] = df2['Application Data'].drop_duplicates(subset='Part #')

# Display the 5 rows
df2['Application Data'].head(5)
```

```python

# Sort the DataFrame df2['Fitment'] by 'Year' in descending order 
df3['Fitment'] = df3['Fitment'].sort_values(by=['Year to', 'Part #'], ascending=[False, True])

# Drop duplicates and keeps the first occurence (which will be the one with the lastest year dye to the previous sorting)
df3['Fitment'] = df3['Fitment'].drop_duplicates(subset='Part #')


# Rename column ['Year to'] to Year
df3['Fitment'].rename(columns={'Year to': 'Year'}, inplace=True)

# Delete the column ['From Year']
df3['Fitment'] = df3['Fitment'].drop('From Year', axis=1)

# Columns new order
df3['Fitment'] = df3['Fitment'][['Part #', 'Year', 'Make', 'Model', 'Style', 'Cab_Length', 'Bed Length', 'Fuel Type', 'Fitment Note' ]]

# Display the 5 rows
df3['Fitment'].head(5)

```

```python
# Define the output file name with the date and time
output_file_name = "Lippert Latest Application Years.xlsx"

# Location path to save the output
file_path = os.path.join(os.path.expanduser("~"), "Desktop", output_file_name)

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(file_path, engine='xlsxwriter')

# Write each dataframe to a different worksheet.
df1['Application Data'].to_excel(writer, sheet_name='Curt Latest Year', index=False, freeze_panes=(1, 0))
df2['Application Data'].to_excel(writer, sheet_name='Aries Latest Year', index=False, freeze_panes=(1, 0))
df3['Fitment'].to_excel(writer, sheet_name='Luverne Latest Year', index=False, freeze_panes=(1, 0))

# Close the Pandas Excel writer and output the Excel file.
writer.close()

# Open the Excel file and set all columns width to 15
with xw.App(visible=False) as app:
    wb = xw.Book(file_path)

    # Loop through all worksheets in the workbook
    for ws in wb.sheets:
        print(f"Processing sheet: {ws.name}")  # Print the name of the sheet being processed

        try:
            # Loop through all columns in the worksheet
            for column in ws.api.UsedRange.Columns:
                column.ColumnWidth = 15
        except Exception as e:
            print(f"An error occurred while processing sheet {ws.name}: {str(e)}")  # Print any errors that occur

    # Save the workbook if needed
    wb.save()

    # Close the workbook
    wb.close()

```
